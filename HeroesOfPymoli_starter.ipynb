### Heroes Of Pymoli Analysis

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
# Count unique player names 
players = purchase_data['SN'].nunique()

# Create and display dataframe of total players
players_dict = [{'Total Players': players}]
players_df = pd.DataFrame(players_dict)
players_df
```

## Purchasing Analysis (Total)

```python
# Find Number of Unique Items
unique_items = len(purchase_data['Item ID'].unique())

# Find Number of Purchases
items_total = purchase_data['Item ID'].count()

# Find Total Revenue
spend_total = purchase_data['Price'].sum()

# Find Average Price
average_price = purchase_data['Price'].mean()

# Create and display summary dataframe
analysis_df = pd.DataFrame([{'Number of Unique Items': unique_items,
                             'Average Price': average_price,
                             'Number of Purchases': items_total,  
                             'Total Revenue': spend_total}])

# Format results
analysis_df['Average Price']= analysis_df['Average Price'].map("${:,.2f}".format)
analysis_df['Total Revenue']= analysis_df['Total Revenue'].map("${:,.2f}".format)
analysis_df
```

## Gender Demographics

```python
# Group by gender
gender_group = purchase_data.groupby('Gender')

# Create dataframe on SN to find numbers of players and mindful of duplicates
gender_num = pd.DataFrame(gender_group['SN'].nunique())
gender_num.rename(columns={'SN':'Total Count'},inplace=True)
```

```python
# Find percentage of each gender
gender_unique = gender_group['SN'].nunique()
gender_per = pd.DataFrame((gender_unique/players)*100)

# Format results
gender_per = gender_per.rename(columns={'SN':'Percentage of Players'})
gender_per['Percentage of Players'] = gender_per['Percentage of Players'].map('{:.2f}%'.format)
```

```python
# Combine gender number and percentage tables
merge_gender = gender_num.merge(gender_per,on='Gender')
merge_gender
```


## Purchasing Analysis (Gender)

```python
# Find Purchase Count
gender_purch_total = pd.DataFrame(gender_group['Item ID'].count())
gender_purch_total.rename(columns={'Item ID':'Purchase Count'},inplace=True)
```

```python
# Find and format Average Purchase Price
gender_purch_avg = pd.DataFrame(gender_group['Price'].mean())
gender_purch_avg.rename(columns={'Price':'Average Purchase Price'},inplace=True)
gender_purch_avg['Average Purchase Price']= gender_purch_avg['Average Purchase Price'].map("${:,.2f}".format)
```

```python
# Find and format Total Purchase Value
gender_purch_tot = pd.DataFrame(gender_group['Price'].sum())
gender_purch_tot.rename(columns={'Price':'Total Purchase Value'},inplace=True)
gender_purch_tot['Total Purchase Value']= gender_purch_tot['Total Purchase Value'].map("${:,.2f}".format)
```

```python
# Find and format Average Total Purchase Per Person
gender_purch_tot_person = pd.DataFrame(gender_group['Price'].sum()/gender_group['Item ID'].count())
gender_purch_tot_person.rename(columns={0:'Avg Total Purchase Per Person'},inplace=True)
gender_purch_tot_person['Avg Total Purchase Per Person']= gender_purch_tot_person['Avg Total Purchase Per Person'].map("${:,.2f}".format)
gender_purch_tot_person

# Combine dataframes
merge_gender = gender_purch_total.merge(gender_purch_avg,on='Gender').merge(gender_purch_tot,on='Gender').merge(gender_purch_tot_person,on='Gender')
merge_gender
```

## Age Demographics

```python
# Create age bins and category names
bins = [0,9,14,19,24,29,34,39,100]
group_names = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Categorize players by age
purchase_data['Age Group'] = pd.cut(purchase_data['Age'],bins,labels=group_names)

# Group by age
age_group = purchase_data.groupby(['Age Group'])

# Count number per age group, mindful of duplicate SNs
age_count = pd.DataFrame(age_group['SN'].nunique())
age_count.rename(columns={'SN':'Total Count'},inplace=True)

# Calculate percent in age group
ages_per = pd.DataFrame((age_group['SN'].nunique()/players)*100)
ages_per.rename(columns={'SN':'Percentage of Players'},inplace=True)
ages_per['Percentage of Players']= ages_per['Percentage of Players'].map("{:.2f}%".format)

# Combine dataframes
merge_ages = age_count.merge(ages_per,on='Age Group')
merge_ages
```

## Purchasing Analysis (Age)

```python
# Find purchase count
age_purch_count = pd.DataFrame(age_group['Purchase ID'].count())
age_purch_count.rename(columns={'Purchase ID':'Purchase Count'},inplace=True)

# Find total purchase value
age_purch_value = pd.DataFrame(age_group['Price'].sum())
age_purch_value.rename(columns={'Price':'Total Purchase Value'},inplace=True)
age_purch_value['Total Purchase Value']= age_purch_value['Total Purchase Value'].map("${:,.2f}".format)

# Find average purchase price
age_purch_avg = pd.DataFrame((age_group['Price'].sum()/age_group['Purchase ID'].count()))
age_purch_avg.rename(columns={0:'Average Purchase Price'},inplace=True)
age_purch_avg['Average Purchase Price']= age_purch_avg['Average Purchase Price'].map("${:,.2f}".format)

# Find average total purchase per person
age_purch_avg_per = pd.DataFrame(age_group['Price'].sum()/age_group['SN'].nunique())
age_purch_avg_per.rename(columns={0:'Avg Total Purchase per Person'},inplace=True)
age_purch_avg_per['Avg Total Purchase per Person']= age_purch_avg_per['Avg Total Purchase per Person'].map("${:,.2f}".format)

#Combine dataframes
merge_age_purch = age_purch_count.merge(age_purch_avg,on='Age Group').merge(age_purch_value,on='Age Group').merge(age_purch_avg_per,on='Age Group')
merge_age_purch
```

## Top Spenders

```python
# Group by player name (spender)
spender_group = purchase_data.groupby(['SN'])

# Find number of purchases for each spender
spender_count = pd.DataFrame(spender_group['Purchase ID'].count())
spender_count.rename(columns={'Purchase ID':'Purchase Count'},inplace=True)

# Find average item purchase price per spender
spender_purch_avg = pd.DataFrame(spender_group['Price'].mean())
spender_purch_avg.rename(columns={'Price':'Average Purchase Price'},inplace=True)

# Find total purchases per spender
spender_purch_tot = pd.DataFrame(spender_group['Price'].sum())
spender_purch_tot.rename(columns={'Price':'Total Purchase Value'},inplace=True)

# Combine dataframes
merged_spender = spender_count.merge(spender_purch_avg,on='SN').merge(spender_purch_tot,on='SN')

# Sort total purchases per spender in descending value
merged_spender = merged_spender.sort_values(['Purchase Count'],ascending=False)

# Format dataframe
merged_spender ['Average Purchase Price']= merged_spender['Average Purchase Price'].map("${:,.2f}".format)
merged_spender ['Total Purchase Value']= merged_spender['Total Purchase Value'].map("${:,.2f}".format)
merged_spender.head()
```

## Most Popular Items

```python
# Group by Item ID and Item Name
popular_group = purchase_data.groupby(['Item ID','Item Name'])

# Find count of purchases by item
popular_count = pd.DataFrame(popular_group['Purchase ID'].count())
popular_count.rename(columns={'Purchase ID':'Purchase Count'},inplace=True)

# Include price for item
popular_item_price = pd.DataFrame(popular_group['Price'].mean())
popular_item_price.rename(columns={'Price':'Item Price'},inplace=True)

# Find total purchase value
popular_purch_value = pd.DataFrame(popular_group['Price'].sum())
popular_purch_value.rename(columns={'Price':'Total Purchase Value'},inplace=True)

# Combine dataframes
merged_popular = popular_count.merge(popular_item_price,on=['Item ID','Item Name']).merge(popular_purch_value,on=['Item ID','Item Name'])

# Sort the purchase count column in descending order
merged_popular2 = merged_popular.sort_values(['Purchase Count'],ascending=False)

# Format the dataframe
merged_popular2 ['Item Price']= merged_popular2['Item Price'].map("${:,.2f}".format)
merged_popular2 ['Total Purchase Value']= merged_popular2['Total Purchase Value'].map("${:,.2f}".format)
merged_popular2.head()
```

## Most Profitable Items

```python
# Sort the above table by total purchase value in descending order
merged_popular = merged_popular.sort_values(['Total Purchase Value'],ascending=False)

# Format the datatable
merged_popular ['Item Price']= merged_popular['Item Price'].map("${:,.2f}".format)
merged_popular ['Total Purchase Value']= merged_popular['Total Purchase Value'].map("${:,.2f}".format)
merged_popular.head()
```
